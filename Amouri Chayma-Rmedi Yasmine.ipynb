```python
# Convertir la colonne 'date' en datetime si ce n'est pas déjà fait
df['date'] = pd.to_datetime(df['date'], format="%d/%m/%Y")

# Trier le DataFrame par date pour assurer une chronologie correcte sur le graphique
df = df.sort_values(by='date')

# Extraire le mois et l'année de la colonne 'date'
df['jour_mois'] = df['date'].dt.strftime('%d/%m')

# Sélectionner uniquement les données tous les 5 jours pour resserrer l'intervalle
df_5_days = df[df.index % 5 == 0]

# Tracer le graphique en courbes
plt.figure(figsize=(10, 6))

plt.plot(df_5_days['jour_mois'], df_5_days['ouverture'], label='Valeur Ouverture', marker='o')
plt.plot(df_5_days['jour_mois'], df_5_days['cloture'], label='Valeur Clôture', marker='o',color='magenta' )

plt.xlabel('Jours')
plt.ylabel('Valeur')
plt.title('Évolution des Valeurs d\'Ouverture et de Clôture au Fil du Temps')
plt.legend()
plt.xticks(rotation=45, ha='right')  # Ajuster la rotation des étiquettes de l'axe x

plt.show()
```

```python
# Trier le DataFrame par date pour assurer une chronologie correcte sur le graphique
df = df.sort_values(by='date')

# Ajouter une colonne de moyenne mobile
df['rolling_mean'] = df['cloture'].rolling(window=5).mean()

# Extraire le jour et le mois de la colonne 'date'
df['jour_mois'] = df['date'].dt.strftime('%d/%m')

# Sélectionner uniquement les données tous les 5 jours pour resserrer l'intervalle
df_5_days = df[df.index % 5 == 0]

# Tracé du graphique en courbes avec rolling mean
plt.figure(figsize=(10, 6))

plt.plot(df_5_days['jour_mois'], df_5_days['cloture'], label='Cotation de Clôture', marker='o')
plt.plot(df_5_days['jour_mois'], df_5_days['rolling_mean'], label='Moyenne Mobile', linestyle='--', color='magenta')

plt.xlabel('Jours')
plt.ylabel('Valeur')
plt.title('Cotations de Clôture avec Moyenne Mobile (DataFrame df)')
plt.legend()
plt.xticks(rotation=45, ha='right')  # Ajuster la rotation des étiquettes de l'axe x

plt.show()


```
